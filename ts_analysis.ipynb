```python
import pandas as pd
import altair as alt
from datetime import datetime
import numpy as np
```

```python
url_mapa = "https://raw.githubusercontent.com/rlrocha/tp3_exemplo/main/Abila.json"
```

```python
df = pd.read_csv('datasets/gps_v2.csv')
df.head()
```

```python
ts_min = '2014-01-06 00:00:00'
ts_max = '2014-01-06 08:59:00'

fmt = '%Y-%m-%d %H:%M:%S'
```

```python
id_list = list(df['id'].unique())

id2_list = []
lat_list = []
long_list = []
t_list = []

for id in id_list:
    df_temp = df.loc[(df['Timestamp'] >= ts_min) & (df['Timestamp'] <= ts_max) & (df['id'] == id)]

    index_list = list(df_temp.index)

    for i in range(np.size(index_list)-1):

        t1 = datetime.strptime(df_temp['Timestamp'][index_list[i]], fmt)
        t2 = datetime.strptime(df_temp['Timestamp'][index_list[i+1]], fmt)

        td = t2-t1

        td_mins = (td.total_seconds()/60)

        if td_mins>=5:
            print('ID: ', id)

            lat1 = df_temp['lat'][index_list[i]]
            lat2 = df_temp['lat'][index_list[i+1]]
            long1 = df_temp['long'][index_list[i]]
            long2 = df_temp['long'][index_list[i+1]]

            id2_list.append(id)
            
            t_list.append(t2)
            
            lat = (lat1 + lat2)/2
            long = (long1 + long2)/2

            lat_list.append(lat)
            long_list.append(long)
            
            print(t1)
            print(t2)
            print(td)
            print('Tempo: ', td_mins)
            print(lat)
            print(long)
```

```python
coord_dict = {'id': id2_list,
              'time': t_list,
              'lat': lat_list,
              'long': long_list}
```

```python
df_coord = pd.DataFrame(coord_dict)
df_coord.head()
```
