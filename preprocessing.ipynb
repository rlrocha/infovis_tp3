### Merge gps.csv e car-assignments.csv e ajuste Timestamp

```python
df1 = pd.read_csv('datasets/gps.csv')
df1.head()
```

```python
df2 = pd.read_csv('datasets/car-assignments.csv')
df2.head()
```

```python
df = pd.merge(df1, df2, left_on='id', right_on='CarID', how='left').drop('CarID', axis=1)
df.head()
```

```python
df["Timestamp"] = pd.to_datetime(df["Timestamp"])
```

```python
df.head()
```

```python
df.to_csv('datasets/car-gps.csv', index=False)
```
